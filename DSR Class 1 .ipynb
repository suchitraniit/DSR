```python
import requests
import re 
url = 'http://www.digbib.org/Johann_Wolfgang_von_Goethe_1749/Faust_I?textonly=1'
file = requests.get(url).text
wordList = file.split()
##print(dir(file))
dict = {}
for word in wordList:
    wordl = word.lower()
    ##print(dir(wordl))
    wordStriped = wordl.replace("@!#$%^&*()[]{};:,./<>?\|`~-=_+", "")
    dict[wordStriped] = wordStriped.count(wordStriped)
print(dict) 
```
